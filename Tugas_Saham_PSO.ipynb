```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Load the dataset
file_path = 'BFIN.csv'
data_historis = pd.read_csv(file_path)

# Rename columns to match expected format
data_historis.rename(columns={
    "Date": "Tanggal",
    "Price": "Terakhir",
    "Open": "Pembukaan",
    "High": "Tertinggi",
    "Low": "Terendah",
    "Change %": "Perubahan%"
}, inplace=True)

# Clean the "Perubahan%" column
if "Perubahan%" in data_historis.columns:
    data_historis["Perubahan%"] = (
        data_historis["Perubahan%"]
        .str.replace('%', '', regex=False)
        .str.replace(',', '.', regex=False)
        .astype(float)
    )

# Clean numeric columns
cols_to_clean = ["Terakhir", "Pembukaan", "Tertinggi", "Terendah"]
for col in cols_to_clean:
    if col in data_historis.columns:
        data_historis[col] = (
            data_historis[col]
            .astype(str)
            .str.replace(',', '', regex=False)
            .astype(float)
        )

# Convert "Tanggal" column to datetime and sort by date
if "Tanggal" in data_historis.columns:
    data_historis["Tanggal"] = pd.to_datetime(data_historis["Tanggal"], format="%d/%m/%Y")
    data_historis.sort_values("Tanggal", inplace=True)

# Calculate log return for analysis
data_historis["Log_Return"] = np.log(data_historis["Terakhir"] / data_historis["Terakhir"].shift(1))

# Basic statistics for log return
mean_log_return = data_historis["Log_Return"].mean()
std_log_return = data_historis["Log_Return"].std()

print(f"Rata-rata log return: {mean_log_return:.6f}")
print(f"Standar deviasi log return: {std_log_return:.6f}")

# Simulation parameters
start_price = data_historis["Terakhir"].iloc[-1]
num_steps = 25  # Simulate for 25 months

# Random walk simulation function
def simulate_random_walk(start_price, mean, std, days, num_simulations):
    simulations = []
    for _ in range(num_simulations):
        prices = [start_price]
        for _ in range(days):
            change = np.random.normal(mean, std)
            next_price = prices[-1] * np.exp(change)
            prices.append(next_price)
        simulations.append(prices)
    return simulations

# Run multiple simulations
num_simulations = 10
simulated_results = simulate_random_walk(start_price, mean_log_return, std_log_return, num_steps, num_simulations)

# Visualize simulation results
colors = cm.rainbow(np.linspace(0, 1, num_simulations))
plt.figure(figsize=(12, 6))
for i, simulation in enumerate(simulated_results):
    plt.plot(range(len(simulation)), simulation, color=colors[i], label=f"Simulasi {i+1}")
plt.plot(range(len(data_historis["Terakhir"])), data_historis["Terakhir"], label="Data Historis", color="black", linewidth=2)
plt.title("Simulasi Pergerakan Saham BFIN Menggunakan Random Walk (25 Bulan)")
plt.xlabel("Bulan")
plt.ylabel("Harga Saham (IDR)")
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))
plt.grid()
plt.show()

# Display the mean and standard deviation after simulation
print("\n=== Statistik Log Return ===")
print(f"Rata-rata log return: {mean_log_return:.6f}")
print(f"Standar deviasi log return: {std_log_return:.6f}")
```
